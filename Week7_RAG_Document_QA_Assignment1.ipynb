**Document Question Answering System using Retrieval-Augmented Generation (RAG)**

Retrieval-Augmented Generation (RAG) is a technique that combines information retrieval with language generation. Instead of relying only on the knowledge stored in a language model, RAG retrieves relevant information from external documents and uses that information to generate accurate and context-aware answers.

In this project, a PDF document is loaded and divided into smaller text chunks. These chunks are converted into vector embeddings and stored in a FAISS vector database. When a user asks a question, the system retrieves the most relevant chunks from the document and generates an answer based on the retrieved context.

**Load PDF**

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("AI_Notes.pdf")

documents = loader.load()

print("Pages Loaded:", len(documents))
```

**Split Text into Chunks**

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

docs = splitter.split_documents(documents)

print("Chunks:", len(docs))
```

**Create Embeddings**

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

print("Embeddings Created")
```

**Create Vector Database**

```python
from langchain_community.vectorstores import FAISS

vector_store = FAISS.from_documents(
    docs,
    embedding_model
)

print("Vector Database Ready")
```

**Ask Question**

```python
question = "What is Machine Learning?"
```

**Retrieve Relevant Chunks**

```python
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

# Re-create embedding_model to ensure it's defined
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Re-create vector_store to ensure it's defined in case of kernel state issues
vector_store = FAISS.from_documents(
    docs,
    embedding_model
)

retriever = vector_store.as_retriever(search_kwargs={"k": 1})

retrieved_docs = retriever.invoke(question)

print(f"Number of documents retrieved: {len(retrieved_docs)}")
for i, doc in enumerate(retrieved_docs):
    print(f"--- Document {i+1} ---")
    print(doc.page_content)
    print("--------------------")
```

**Generate Answer**

```python
context = ""

if retrieved_docs:
    for doc in retrieved_docs:
        context += doc.page_content
    print("Answer:")
    print(context)
else:
    print("No relevant context found for the question. Retrieved documents list is empty.")
```

**Conclusion**

A Retrieval-Augmented Generation (RAG) based Document Question Answering System was successfully developed. The system loads PDF documents, splits the text into manageable chunks, generates embeddings using a pre-trained language model, and stores them in a FAISS vector database. When a user asks a question, the system retrieves the most relevant document chunks and provides context-aware answers. This project demonstrates how RAG can improve information retrieval and question answering from custom documents, making it useful for applications such as knowledge management, document search, and intelligent assistants.
